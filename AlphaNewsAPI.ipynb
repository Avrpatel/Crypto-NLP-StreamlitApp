```python
# symbol = "CRYPTO:BTC"
#t_from = "20200809T0000"
#t_to = "20201008T0000"
# &time_from=20200809T0000&time_to=20201008T0000

def retrieve_data():
    url = f"https://www.alphavantage.co/query?function=NEWS_SENTIMENT&tickers=CRYPTO:BNB&time_from=20220701T0000&time_to=20220731T0000&sort=EARLIEST&limit=200&apikey={api_key}"
    response = requests.get(url).json()
    return response
```

```python
data=retrieve_data()
#print(data)
```

```python
df = pd.DataFrame.from_dict(data)
df1 = df["feed"].apply(pd.Series)
df1["time_published"] = pd.to_datetime(df1["time_published"])
display(df1.head())
#display(df1.tail())
```

```python
filepath = Path('bnb_july.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
df1.to_csv(filepath) 
```
