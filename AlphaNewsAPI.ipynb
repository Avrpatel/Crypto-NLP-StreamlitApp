```python
def retrieve_data():
    api_key = "api_key"
    symbol = "CRYPTO:BTC"
    time_from = "20210410T0130" #format 20220410T0130
    time_to = "20220410T0130"
    url = f"https://www.alphavantage.co/query?function=NEWS_SENTIMENT&tickers={symbol}&time_from={time_from}&time_to={time_to}&apikey={api_key}"
    response = requests.get(url).json()
    return response
```

```python
data=retrieve_data()
#print(data)
```

```python
df = pd.DataFrame.from_dict(data)
df1 = df["feed"].apply(pd.Series)
df1["time_published"] = pd.to_datetime(df1["time_published"])
df1.head()
```

```python
filepath = Path('btc.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
df1.to_csv(filepath) 
```
